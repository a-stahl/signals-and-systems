```python
# Plot for the power spectral denstiy of a signal
f, Pxx_den = signal.periodogram(funky, fs)
plt.plot(f, 10*np.log10(Pxx_den))
plt.grid()
plt.ylabel('Power Spectrum (dB)')
plt.xlabel('Frequency')
plt.show()
```

### Filter design using signal.firwin
https://docs.scipy.org/doc/scipy-0.14.0/reference/generated/scipy.signal.butter.html

```python
f_cutoff = 200  #Required cut-off frequency of the filter 
fNorm = f_cutoff/(fs/2) #The cut-off must be normalized for digital filter design
N=10             #Order of filter

b_low_fir = signal.firwin(N, fNorm) # Generate filter coeficients (numerator)
a_low_fir = np.ones(1)              # Denominator, 1 for this case

#Magnitude response of the filter
w_low_fir, h_low_fir = signal.freqz(b_low_fir,a_low_fir) #f=radians per sample
fig, ax1 = plt.subplots(figsize=(5,3)) 
ax1.plot(w_low_fir,np.abs(h_low_fir),color='blue') 
ax1.set_ylabel('Filter Magnitude (dB)',color='blue') 
ax1.set_xlabel('Frequency (radians/sample)')
ax1.grid()
plt.show()

# Filtering the audio file
funkylow_fir = signal.filtfilt(b_low_fir,a_low_fir,funky)

#Power-spectrum
f_low_fir, Pxx_den_low_fir = signal.periodogram(funkylow_fir, fs)
plt.plot(f_low_fir, 10*np.log10(Pxx_den_low_fir))
plt.grid()
plt.ylabel('Power Spectrum (dB)')
plt.xlabel('Frequency')
plt.show()
```

```python
## Pole-zero plot
H_low_fir = signal.lti(b_low_fir,a_low_fir)
theta = np.linspace(0,2*np.pi,101)
plt.figure()
plt.plot(np.cos(theta), np.sin(theta),'lightgrey')
plt.plot(np.real(H_low_fir.zeros),np.imag(H_low_fir.zeros),'ko',
        markerfacecolor='none')
plt.plot(np.real(H_low_fir.poles),np.imag(H_low_fir.poles),'kx',
        markerfacecolor='none')
plt.axes().set_aspect('equal')
plt.show()
```

### Butterworth Filter

```python
f_cutoff = 7000
fNorm = f_cutoff/(fs/2)
N=5
b_low_fir, a_low_fir = signal.butter(N, fNorm, 'highpass')
# Use the third argument to switch between lowpass, highpass, bandpass etc.

funkylow = signal.filtfilt(b_low_fir,a_low_fir,funky)

w_low_fir, h_low_fir = signal.freqz(b_low_fir,a_low_fir) #f=radians per sample
fig, ax1 = plt.subplots(figsize=(5,3)) 
ax1.plot(w_low_fir,np.abs(h_low_fir),color='blue') 
ax1.set_ylabel('Filter Magnitude (dB)',color='blue') 
ax1.set_xlabel('Frequency (radians/sample)')
ax1.grid()
plt.show()

f_low_fir, Pxx_den_low_fir = signal.periodogram(funkylow, fs)
plt.plot(f_low_fir, 10*np.log10(Pxx_den_low_fir))
plt.grid()
plt.ylabel('Power Spectrum (dB)')
plt.xlabel('Frequency')
plt.show()
```

### Task 2: Filter Design

1) Design a low-pass butterworth filter of order 20 and cutoff frequency 1500Hz. Show the pole-zero, the magnitude respons and the power-spectrum of the filtered output.

2) Design a low-pass filter using firwin of order 20 and cutoff frequency 1500Hz. Show the pole-zero, the magnitude respons and the power-spectrum of the filtered output.

3) Design a low-pass filter using firwin of order 200 and cutoff frequency 1500Hz. Show the pole-zero, the magnitude respons and the power-spectrum of the filtered output.

4) Compare the results of the 3 filters designed above.

```python
# Filter 1

f_cutoff = 1500
fNorm = f_cutoff/(fs/2)
N=20 # order of filter
b_low_fir, a_low_fir = signal.butter(N, fNorm, 'lowpass')
# Use the third argument to switch between lowpass, highpass, bandpass etc.

funkylow = signal.filtfilt(b_low_fir,a_low_fir,funky)

w_low_fir, h_low_fir = signal.freqz(b_low_fir,a_low_fir) #f=radians per sample
fig, ax1 = plt.subplots(figsize=(5,3)) 
ax1.plot(w_low_fir,np.abs(h_low_fir),color='blue') 
ax1.set_ylabel('Filter Magnitude (dB)',color='blue') 
ax1.set_xlabel('Frequency (radians/sample)')
ax1.grid()
plt.show()

f_low_fir, Pxx_den_low_fir = signal.periodogram(funkylow, fs)
plt.plot(f_low_fir, 10*np.log10(Pxx_den_low_fir))
plt.grid()
plt.ylabel('Power Spectrum (dB)')
plt.xlabel('Frequency')
plt.show()

## Pole-zero plot
H_low_fir = signal.lti(b_low_fir,a_low_fir)
theta = np.linspace(0,2*np.pi,101)
plt.figure()
plt.plot(np.cos(theta), np.sin(theta),'lightgrey')
plt.plot(np.real(H_low_fir.zeros),np.imag(H_low_fir.zeros),'ko',
        markerfacecolor='none')
plt.plot(np.real(H_low_fir.poles),np.imag(H_low_fir.poles),'kx',
        markerfacecolor='none')
plt.axes().set_aspect('equal')
plt.show()
```

```python
# Filter 2

f_cutoff = 1500  #Required cut-off frequency of the filter 
fNorm = f_cutoff/(fs/2) #The cut-off must be normalized for digital filter design
N=20             #Order of filter

b_low_fir = signal.firwin(N, fNorm) # Generate filter coeficients (numerator)
a_low_fir = np.ones(1)              # Denominator, 1 for this case

#Magnitude response of the filter
w_low_fir, h_low_fir = signal.freqz(b_low_fir,a_low_fir) #f=radians per sample
fig, ax1 = plt.subplots(figsize=(5,3)) 
ax1.plot(w_low_fir,np.abs(h_low_fir),color='blue') 
ax1.set_ylabel('Filter Magnitude (dB)',color='blue') 
ax1.set_xlabel('Frequency (radians/sample)')
ax1.grid()
plt.show()

# Filtering the audio file
funkylow_fir = signal.filtfilt(b_low_fir,a_low_fir,funky)

#Power-spectrum
f_low_fir, Pxx_den_low_fir = signal.periodogram(funkylow_fir, fs)
plt.plot(f_low_fir, 10*np.log10(Pxx_den_low_fir))
plt.grid()
plt.ylabel('Power Spectrum (dB)')
plt.xlabel('Frequency')
plt.show()

## Pole-zero plot
H_low_fir = signal.lti(b_low_fir,a_low_fir)
theta = np.linspace(0,2*np.pi,101)
plt.figure()
plt.plot(np.cos(theta), np.sin(theta),'lightgrey')
plt.plot(np.real(H_low_fir.zeros),np.imag(H_low_fir.zeros),'ko',
        markerfacecolor='none')
plt.plot(np.real(H_low_fir.poles),np.imag(H_low_fir.poles),'kx',
        markerfacecolor='none')
plt.axes().set_aspect('equal')
plt.show()
```

```python
# Filter 3

f_cutoff = 1500  #Required cut-off frequency of the filter 
fNorm = f_cutoff/(fs/2) #The cut-off must be normalized for digital filter design
N=2000             #Order of filter

b_low_fir = signal.firwin(N, fNorm) # Generate filter coeficients (numerator)
a_low_fir = np.ones(1)              # Denominator, 1 for this case

#Magnitude response of the filter
w_low_fir, h_low_fir = signal.freqz(b_low_fir,a_low_fir) #f=radians per sample
fig, ax1 = plt.subplots(figsize=(5,3)) 
ax1.plot(w_low_fir,np.abs(h_low_fir),color='blue') 
ax1.set_ylabel('Filter Magnitude (dB)',color='blue') 
ax1.set_xlabel('Frequency (radians/sample)')
ax1.grid()
plt.show()

# Filtering the audio file
funkylow_fir = signal.filtfilt(b_low_fir,a_low_fir,funky)

#Power-spectrum
f_low_fir, Pxx_den_low_fir = signal.periodogram(funkylow_fir, fs)
plt.plot(f_low_fir, 10*np.log10(Pxx_den_low_fir))
plt.grid()
plt.ylabel('Power Spectrum (dB)')
plt.xlabel('Frequency')
plt.show()

## Pole-zero plot
H_low_fir = signal.lti(b_low_fir,a_low_fir)
theta = np.linspace(0,2*np.pi,101)
plt.figure(figsize=(16,16))
plt.plot(np.cos(theta), np.sin(theta),'lightgrey')
plt.plot(np.real(H_low_fir.zeros),np.imag(H_low_fir.zeros),'ko',
        markerfacecolor='none')
plt.plot(np.real(H_low_fir.poles),np.imag(H_low_fir.poles),'kx',
        markerfacecolor='none')
plt.axes().set_aspect('equal')
plt.show()
```

## Comments

Obviously the FIR filters do not have poles in their pole zero plot because otherwise they would not be finite impulse response. What interested me was how the butterworth filter had some of its zeros located outside of the unit circle, but for the most part the FIR filters only had one or two. For the FIR filters we can see that as the order increased the filter quality increased, but at the cost of implementation complexity and hardware cost. In the power spectrum frequency plot of the lower order filter FIR we can also visually see the zeros, but not in the butterworth or 2000 order FIR. If I were to implement one of these for a simple project I would probably choose the order 20 plot for its stability and cost effectiveness.
