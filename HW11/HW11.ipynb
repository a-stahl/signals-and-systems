```python
# import libraries
import matplotlib.pyplot as plt
import numpy as np
from scipy import signal
```

```python
# plot the magnitude and phase response

num = [1, -0.6, -0.35, 0.9]
den = [1, -1, 0.49, -0.49]

w, h = signal.freqz(num, den)
plt.subplot(2,1,1)
plt.title('Question Five Magnitude and Phase Response')
plt.plot(w, 20 * np.log10(np.real(abs(h))), 'b')
plt.ylabel('Amplitude [dB]', color='b')
plt.xlabel('Frequency [rad/sample]')
plt.subplot(2,1,2)
plt.plot(w, np.angle(h), 'r')
plt.ylabel('Angle (rad)', color='b')
plt.show()


#ignoring the div/0 error here
```

```python
# create the pole-zero plot

H = signal.lti(num, den)
theta = np.linspace(0,2*np.pi,101)
plt.plot(np.cos(theta), np.sin(theta), 'lightgrey')
plt.plot(np.real(H.zeros), np.imag(H.zeros), 'ko', markerfacecolor='none')
plt.plot(np.real(H.poles), np.imag(H.poles), 'kx')
plt.axes().set_aspect('equal')
```
